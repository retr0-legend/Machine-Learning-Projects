### Import directories 

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
le = LabelEncoder()
```

# Load dataset

```python
tayyba = pd.read_csv("heart_disease.csv")
```

## Drop irralivant columns 

```python
tayyba = tayyba.drop(['Country'], axis=1)
```

```python
tayyba.columns
```

```python
tayyba.shape
```

##  Convert all categorical Columns

```python
for c in tayyba.columns:
    if tayyba[c].dtype == 'object' :  
        tayyba[c] = le.fit_transform(tayyba[c])
```

```python
tayyba.info()
```

## Find Correlation of dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
correlation = tayyba.corr()
print(correlation)
```

## Heat Map of df Correlation

```python
plt.figure(figsize=(46,23))
sns.heatmap(correlation, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

## Regression Split

```python
X =tayyba.drop('Blood_Pressure', axis=1)
Y = tayyba['Blood_Pressure']
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import (mean_absolute_error, mean_squared_error, 
r2_score)
```

```python
X_TRAIN,X_TEST,Y_TRAIN,Y_TEST=train_test_split(X,Y,test_size=0.25,random_state=25)
print("Size of Train X=",len(X_TRAIN))
print("Size of Train Y=",len(Y_TRAIN))
print("Size of Test X=",len(X_TEST))
print("Size of Test Y=",len(Y_TEST))
```

 # 1) Linear Regression

## Prediction on Training Data

```python
from sklearn.linear_model import LinearRegression
```

```python
tayyba = LinearRegression()
```

```python
tayyba.fit(X_TRAIN, Y_TRAIN)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
training_data_prediction = tayyba.predict(X_TRAIN)
print(training_data_prediction)
```

### Accuracy

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(Y_TRAIN, training_data_prediction)
```

```python
print("R-Squared:", r2)
```

### Mean Absolut error

```python
from sklearn.metrics import mean_absolute_error
from sklearn import metrics
```

```python
score_2 = metrics.mean_absolute_error(Y_TRAIN, training_data_prediction)
```

```python
print(len(Y_TRAIN))
print(len(training_data_prediction))
```

```python
print("R-Squared:", r2)
print("Mean Absolute Error:", score_2)
```

## Graphical Represention 

```python
plt.scatter(Y_TRAIN, training_data_prediction)
plt.xlabel("Actual Blood Pressure")
plt.ylabel("Predicted Blood Pressure")
plt.title("Actual Blood Pressure vs Preicted Blood Pressure")
plt.show()
```

# Prediction on Test Data

```python
y_pred = tayyba.predict(X_TEST)
y_pred
print(y_pred)
```

### Acuraccy

```python
score_1 = metrics.r2_score(Y_TEST, y_pred)
```

```python
score_2 = metrics.mean_absolute_error(Y_TEST, y_pred)
```

```python
print("R squared : ", score_1)
print('Mean Absolute Error : ', score_2)
```

## Graphical Representaion 

```python
plt.scatter(Y_TEST, y_pred)
plt.xlabel("Actual Blood Pressure")
plt.ylabel("Predicted Blood Pressure")
plt.title("Actual Blood Pressure vs Preicted Blood Pressure")
plt.show()
```
